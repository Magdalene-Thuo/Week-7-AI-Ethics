```python
import matplotlib.pyplot as plt

# Sample values for visualization
disparate_impact = 0.65  # less than 0.8 indicates bias
fpr_difference = 0.18    # difference in false positive rates between groups

# Create bar chart
labels = ['Disparate Impact', 'FPR Difference']
values = [disparate_impact, fpr_difference]
colors = ['crimson' if val < 0.8 else 'green' for val in values]

plt.figure(figsize=(8, 5))
bars = plt.bar(labels, values, color=colors)
plt.axhline(0.8, color='gray', linestyle='--', linewidth=1, label='Fairness Threshold (0.8)')
plt.axhline(0.0, color='black', linewidth=0.5)

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.02, f'{yval:.2f}', ha='center', va='bottom')

plt.title('Bias Metrics Visualization - COMPAS Dataset')
plt.ylabel('Metric Value')
plt.ylim(0, 1.2)
plt.legend()
plt.tight_layout()
plt.show()
```
